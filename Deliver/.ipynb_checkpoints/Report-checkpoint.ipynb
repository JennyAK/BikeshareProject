## Objective

This Bikeshare data set consist of two data sets, 'daily' and 'hourly' data. Both data sets contains data on day, season, year, month, hoilday, weekday, workingday, weatherday, temperature, feeling temperature, humidity, windspeed, causual rental, registered rental and count of rentals. The hourly data set also includes hour.

The daily data set had 731 rows and 16 columns. The hourly data set had 17,379 rows and 17 columns. Because of the difference in the number of rows, analysis were done on both data sets.

This projects aims to predict the count of bike rentals using the features in these data sets.

<a id="clean"></a>
## Preprocessing the Data

Before doing analysis, the data sets had to be preprocessed.

**Linear Models**

First, features that did not seem to be useful were removed. From each data set, the features (X) were extracted and the predicting column (y) were seperated. Then the data sets were split into a trainning set and test set, giving us a total of 4 data sets. All data set were normailzed according to the x_train data.

**Deep Neural Net Models**

Preprocessing for the DNN models were more involved. First, repeative and useless features were removed ('instant', 'dteday', 'casual', and 'registered'). Next the data types of the features were examined to see which were actually categorical data. Categorical data were converted to dummy features, using *get_dummies* from pandas.

After dummifying categorical data, the number of features grew to 37 for Daily data and 38 for Hourly data. Instead of using all features, *ExtraTreesClassifier* from sklearns, was used to determine the top ten features.

Both data sets were normalized with *MinMaxScaler* from sklearns. Then, finally both data sets were split into train and test using train_test_split from sklearns. 80% of the data was used for trainning and 20% was used for test.

# Linear Model

Our first model (in the EDA phase) used Linear Regression(s). First multiple linear regression was used to see how all the kept features affected the model. The daily data set had a high score of **80.5%**. The hourly data set gave a poor score of **33.1%**.

To see how the highest correlated feature did to predict the count, a correlation was done to see what was the highest correlated feature to the count. The daily data set had 'atemp' (feeling temperture) and hourly data had
'temp' (tempearture) as the higher correlated features. Next a simple linear regression was done using the respective features.

## Deep Neural Net
The second model to predict total count of bike rentals was a deep neural network. There were two layers with a 'relu' activation. The output layer had a 'linear' activation because the model is predicting a float value.

When compiling the model, the loss was set to mean squared error. The optimizer was left as the default (RMSprop). The metrics were accuracy and mean squared error.

The batch size was 100, and the model ran for 50 epochs.

Both Daily and Hourly data sets had the same paramenters when building the deep neural network.

### Daily:



When evaluating the Daily test set, the mean squared error was 0.007.

### Hourly:



When evaluating the Hourly test set, the mean squared error was 0.013.

<a id="conclusion"></a>
## Conclusion

The multiple linear regression and simple linear regression showed that the Daily data set gave a better prediction for the total count of bike rentals. Both DNN models had very low mean squared errors. However the Daily data set was almost twice as low as the Hourly data set.

Over all the Daily data set will give a better prediction of the total count of bike rentals. Using a DNN model will have a very low mean squared error of both, but the Hourly data set will give a mean squared error that is almost double of the Daily data set.
