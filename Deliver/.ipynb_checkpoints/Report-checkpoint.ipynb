## Objective

This Bikeshare data set consist of two data sets, 'daily' and 'hourly' data. Both data sets contains data on day, season, year, month, hoilday, weekday, workingday, weatherday, temperature, average temperature, humidity, windspeed, causual rental, registered rental and count of rentals. The hourly data set also includes hour.

This projects aims to predict the count of bike rentals using the features in this data set.

<a id="clean"></a>
## Preprocessing the Data

Before doing analysis, the data sets had to be preprocessed. First, features that did not seem to be useful were removed. From each data set, the features (X) were extracted and the predicting column (y) were seperated. Then the data sets were split into a trainning set and test set, giving us a total of 4 data sets. All data set were normailzed according to the x_train data.

## Naive Bayes

In the second model, naive bayes was used to first look at which feature(s) whould best predict the bike rental counts, given a certain accuracy rate. Because of the oversight with the 'casual' and 'registered' features, this naive bayes model was built so we can give a accuracy rate we want to achive, and the model will give us the freature(s) that we let us achive that accuracy rate. 

For both Daily and Hourly data sets, only a single feature was needed to predict the total bike rental count. Both data sets were set to predict with an 80% accuracy and both data sets gave 'atemp' (average temperature) as the feature needed to acheive that accuracy rate. With Naive Bayes, both Daily and Hourly datasets were givin to predict total bike rental counts with a 98% and above accuracy rate.
